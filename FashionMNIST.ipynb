```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Flatten, Dense
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.datasets import fashion_mnist
(x_train, x_lab),(y_test, y_lab) = fashion_mnist.load_data()
```

```python
x_train.shape
```

```python
print('Train: X=%s, y=%s' % (x_train.shape, x_lab.shape))
print('Test: X=%s, y=%s' % (y_test.shape, y_lab.shape))
# plot first few images
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# plot raw pixel data
	plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
# show the figure
plt.show()
```

```python
plt.imshow(x_train[0])
plt.title('Class: {}'.format(x_lab[0]))
plt.figure()
```

```python
model = Sequential()
model.add(Flatten(input_shape=((28,28))))
model.add(Dense(200, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

```python
model.compile(optimizer = "adam", loss="sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
# Fitting the Model
model.fit(x_train, x_lab, epochs = 20)
```

```python
model.evaluate(y_test, y_lab)
```

```python
# Predict the first 10 images, Probability Distribution
p = model.predict(y_test[:10])
print(p)
```

```python
pred = np.argmax(p, axis=1)
print(pred)
print(y_lab[:10])
```

```python
# Visualizing the result
for i in range(10):
  plt.imshow(y_test[i], cmap="binary")
  plt.title('Original: {}, Predicted: {}'.format(y_lab[i], pred[i]))
  plt.axis("Off")
  plt.figure()
```
